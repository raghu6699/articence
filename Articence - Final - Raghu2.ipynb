plot profitability rather than profits. (mean profitability or total profits/total cost)



```python
df_gb = pd.concat([df.groupby("Item Type")["Total Cost"].sum(),df.groupby("Item Type")["Total Profit"].sum()],axis=1)
df_gb["profitability"] = df_gb["Total Profit"]/df_gb["Total Cost"]
df_gb
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x=df_gb.index,y=df_gb.profitability)
```

# # Question:2


#### Which products are the easiest to sell. (hint: look at the ship date and order date) by region. Is there a relationship between products or region?

First we will calculate the date difference and mean it to find out the average delivery duration.

```python
df1 = df[['Region', 'Item Type','Order Date', 'Ship Date']]
df1['Order Date'] = pd.to_datetime(df1['Order Date'])
df1['Ship Date'] = pd.to_datetime(df1['Ship Date'])
df1["Date_Difference"] = (df1['Ship Date'] - df1['Order Date']).dt.days
df2 = df1.groupby(['Item Type', 'Region'])['Date_Difference'].mean().reset_index()
df2.head()
```

```python
df3=pd.crosstab(df2['Region'], df2['Item Type'], values=df2['Date_Difference'], aggfunc='mean').round(2)
df3
```

```python
import seaborn as sns
cm = sns.light_palette("green", as_cmap=True)
s = df3.style.background_gradient(cmap=cm)
s
```
